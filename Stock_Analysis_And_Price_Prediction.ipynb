# Data Science Project - Stock Market Analysis and Price Prediction using LSTM

## 1. Stock Market Analysis 

In this section, we will be doing some basic analysis of the stocks from 4 tech companies: Apple, Google, Microsoft and Amazon. 

```python
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
```

First, we will be using the yfinance library to read the stock data. By specifying the start time and end time, we can extract the stock data from a given period. 

```python
stock_list = ['AAPL', 'GOOG', 'MSFT', 'AMZN']
data_list = [None] * len(stock_list)

##set the time period to one year ending in 2024/05/01
end = datetime(2024, 5, 1)
start = datetime(end.year - 1, end.month, end.day)

for i, stock in enumerate(stock_list):   
    data_list[i] = yf.download(stock, start=start, end=end)
```

```python
company_list = ["APPLE", "GOOGLE", "MICROSOFT", "AMAZON"]

for data, name in zip(data_list, company_list):
    data["Company"] = name
    
df = pd.concat(data_list, axis=0)
df.sort_index(inplace=True)
df.head(10)
```

The extracted stock data contain the open price, high price, low price, close price, adjusted close price, and volume of the stocks. The close price simply refers to the cost of shares at the end of the day, the adjusted close price takes dividends, stock splits, and new stock offerings into account. Therefore, the adjusted close price is a more accurate indicator of stock value since it starts where the close price finishes. In this analysis, we will be mainly using the adjusted close price.

Next, we check if the data have missing values

```python
df.info()
```

Now, let's visualize the change of adjusted closing price of each stock in the past year

```python
for i, data in enumerate(data_list):
    fig = plt.subplots(figsize=(20, 4))
    sns.lineplot(data=data, x="Date", y="Adj Close")
    plt.title(f"Adjusted Closing Price ({stock_list[i]})")
```

The plots above indicate that Apple, Google, and Microsoft stocks have demonstrated a consistent upward trend over the past year, characterized by minimal volatility. Conversely, Amazon's stock has exhibited greater variability and fluctuations during the same period, displaying a weaker performance compared to its counterparts.

The change of volumes provides insights into market dynamics and investor behavior. For example, sudden spikes in volume may signal increased interest or activity in a particular stock, possibly due to significant news events, earnings announcements, or changes in market sentiment. Conversely, periods of low volume may suggest decreased interest or reduced trading activity, which could indicate a lack of conviction or uncertainty among investors. Let's also plot the change of volumes of the stocks.

```python
for i, data in enumerate(data_list):
    fig = plt.subplots(figsize=(20, 4))
    sns.lineplot(data=data, x="Date", y="Volume")
    plt.title(f"Volume ({stock_list[i]})")
```

We see consistent fluctuations and many spikes along the year, seems like the some spikes occurred simultaneously across all 4 stocks. To verify this, we can plot the change of volumes in one graph so that we can see it clearly. 

```python
fig = plt.subplots(figsize=(20, 6))

for i, data in enumerate(data_list):
    plt.plot(data['Volume'], label=stock_list[i]) 
    plt.legend() 

plt.title(f"Volume of All 4 Stocks")
plt.show()
```

Now it is clear that many spikes occurred simultaneously which indicates similar volume fluctuations across these tech companies. This can be attributed to shared sector trends and market conditions. As technology sector leaders, these companies often move in tandem due to common industry dynamics and broader market sentiment. During periods of sector-wide news, events, or shifts in market sentiment, investors may adjust their trading activity across multiple tech stocks simultaneously, leading to synchronized volume patterns. 

One important indicator used in stock market analysis is the moving average(MA) indicator, MA offers the advantage of smoothing price data, confirming price movements and identifying trends. By averaging prices over a specific time period, moving averages help to filter out short-term noise and highlight the underlying trend, enabling traders and investors to make more informed decisions about entry and exit points, as well as managing risk effectively.

Below we use time periods of 10 days, 20 days and 50 days to calcualte the MA closing price and visualize each of the prices.

```python
ma_periods = [10, 20, 50]

for ma in ma_periods:
    for data in data_list:
        column = f"{ma}-day Moving Average"
        data[column] = data['Adj Close'].rolling(ma).mean()
        

fig, axes = plt.subplots(nrows=2, ncols=2)
fig.set_figheight(8)
fig.set_figwidth(15)
for i, data in enumerate(data_list):
    data[['Adj Close', '10-day Moving Average', '20-day Moving Average', '50-day Moving Average']].plot(kind='line', ax=axes[i//2, i%2])
    axes[i//2, i%2].set_title(f'{stock_list[i]}')
fig.tight_layout()
```

Next, we will be exploring the daily return of stocks. The daily return of a stock is the percentage change in its price from one trading day to the next. Daily returns are useful because they provide valuable insights into the performance and volatility of a stock over short-term periods. They allow investors and traders to assess the magnitude and direction of price changes on a daily basis, helping them to evaluate investment strategies and manage risks.

```python
##calcuate daily returns 
for data in data_list:
    data['Daily Return'] = data['Adj Close'].pct_change()
    
fig, axes = plt.subplots(nrows=2, ncols=2)
fig.set_figheight(8)
fig.set_figwidth(15)
for i, data in enumerate(data_list):
    data['Daily Return'].plot(kind='line', marker='.', ax=axes[i//2, i%2])
    axes[i//2, i%2].legend(loc='upper right') 
    axes[i//2, i%2].set_title(f'{stock_list[i]}')
fig.tight_layout()
```

Despite having some outliers in each plot, the level of fluctuations in daily return fall within a small range which indicates a moderate level of volatility in the stock's price movements. 

Another analysis we will be doing here is to explore the correlations between different stocks closing prices and daily returns. Stocks within the same industry or sector often exhibit higher correlations. As tech leaders, it is likely that there will be correlations between these companies.

```python
close_df = yf.download(stock_list, start=start, end=end)['Adj Close']
close_df.head()
```

```python
returns_df = close_df.pct_change()
returns_df.head()
```

Let's plot the pair correlation between closing price of each stock using scatter plots.

```python
sns.pairplot(close_df, height=2)
```

We can see from the pair plot that there is a positive correlation between the closing price of Amazon, Google and Microsoft stocks, while there is no obvious correlation between the Apple stock and other stocks. 

Now the pair correlation between daily return of each stock.

```python
sns.pairplot(returns_df.dropna(), height=2)
```

There seems to be a positive correlation between the daily return of all 4 stocks but the correlation is weaker than the closing price correlation.

Alternatively, we can also show the correlations by using the heatmap. 

```python
sns.heatmap(close_df.corr(), 
            annot=True, 
            cmap='BuPu')
plt.title("Correlation (Closing Price)")
```

```python
corr = returns_df.dropna().corr()
sns.heatmap(corr,
            xticklabels=corr.columns,
            yticklabels=corr.columns, 
            annot=True,
            cmap='BuPu')
plt.title("Correlation (Daily Return)")
```

Great! Just like we suspected in the pair plot there is a strong correlation between the closing price of Amazon, Google and Microsoft. If we recall the plot of the closing price of these 3 stocks over the past year they both demonstrated similar upward trends whereas the closing price of Apple stock does not. As for the daily return, there is positive correlations among all companies stocks, albeit being relatively weaker.

Our final analysis is the risk analysis. One fundamental method to quantify risk involves comparing the expected return to the standard deviation of daily returns, using the information gathered on daily percentage returns.

```python
rets = returns_df.dropna()
plt.figure(figsize=(10,4))
plt.scatter(rets.mean(),rets.std(),s=30)
plt.xlabel('Expected Return')
plt.ylabel('Risk')

#Adding annotations in the scatter plot
for name, x, y in zip(rets.columns, rets.mean(), rets.std()):
    plt.annotate(
        name,
        xy=(x,y), xytext=(-120,20),
        textcoords = 'offset points', 
        ha = 'right', 
        va = 'bottom',
        arrowprops = dict(arrowstyle='->'))
```

## 2. Stock Price Prediction 

In this section, we will try to predict the closing price of Google's stock by using LSTM.

First we extract the data from the past three years. 

```python
end = datetime(2024, 5, 1)
start = datetime(end.year - 3, end.month, end.day)
google_df = yf.download('GOOG', start=start, end=end)
google_df.head()
```

Split the data into 80% train data and 20% test data

```python
data = google_df['Close'].values
train_split = 0.8
test_split = 1 - train_split
train_len = int(np.ceil(len(data) * 0.8))
print(f"Train length: {train_len}")
print(f"Test length: {len(data) - train_len}")
```

Scale the data using a MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler

data = np.reshape(data, (len(data), 1))
scaler = MinMaxScaler(feature_range=(0,1))
scaled_data = scaler.fit_transform(data)
```

Now we need to transform our data. Firstly we need to define a variable called num_of_states, it specifies how many past states(past prices) we want in order to predict the next price. Here, num_of_states is set to 30 days. Then we generate the data by constantly taking every 30 states as features and the following price as label.  

```python
##we need 30 past prices to predict the next price
num_of_states = 30
train_data = scaled_data[0:train_len, :]
x_train = []
y_train = []

##generate train data 
for i in range(num_of_states, train_len):
    x_train.append(train_data[i-num_of_states:i, 0])
    y_train.append(train_data[i, 0])

x_train, y_train = np.array(x_train), np.array(y_train)
x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))
print(x_train.shape)
print(y_train.shape)
```

Next we build our neural network for prediction, which contains two layers of LSTM units a one Dense layer.

```python
from keras.models import Sequential, Model
from keras.layers import Dense, LSTM

model = Sequential()
model.add(LSTM(32, return_sequences=True, input_shape=(x_train.shape[1], 1)))
model.add(LSTM(32, return_sequences= False))
model.add(Dense(64))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mean_squared_error')
```

Train the data by 30 epochs with a batch size of 32.

```python
model.fit(x_train, y_train, batch_size=32, epochs=30)
```

Transform and generate the test data like we did with the train data earlier.

```python
test_data = scaled_data[train_len-num_of_states:, :]
x_test = []
y_test = data[train_len:, :]
for i in range(num_of_states, len(test_data)):
    x_test.append(test_data[i-num_of_states:i, 0])
    
x_test = np.array(x_test)
x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], 1 ))
```

Predict the test data and then scale it back

```python
y_pred = model.predict(x_test)
y_pred = scaler.inverse_transform(y_pred)
```

```python
rmse = np.sqrt(np.mean(((y_pred - y_test) ** 2)))
print(f"rmse: {rmse}")
```

Visualize the results, compare our predictions to the true values.

```python
train = google_df.iloc[:train_len, :]
test = google_df.iloc[train_len:, :]
test['Prediction'] = y_pred[:, 0]

# Visualize the data
plt.figure(figsize=(12,6))
plt.title('Model prediction vs True values')
plt.xlabel('Date', fontsize=12)
plt.ylabel('Close Price USD ($)', fontsize=12)
plt.plot(train['Close'])
plt.plot(test[['Close', 'Prediction']])
plt.legend(['Train', 'True', 'Prediction'], loc='lower right')
plt.show()
```
